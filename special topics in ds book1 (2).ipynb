# Scatterplot for Canada ,Costa Rica, Israel , Mexico ,India

```python
import seaborn as sns
import matplotlib.pyplot as plt

def scatter_country_data(countries, df):
    plt.figure(figsize=(12, 7))
    sns.set(style="whitegrid")
    
    # Define unique colors and markers for each country
    colors = sns.color_palette("husl", len(countries))
    markers = ['o', 's', 'D', '^', 'P']  # Different marker styles

    # Plot data for each country
    for i, country in enumerate(countries):
        country_data = df[df['Geography 1/'] == country]
        
        if not country_data.empty:
            years = country_data.columns[2:-1]  # Extract the years
            values = country_data.iloc[0, 2:-1].replace(',', '', regex=True).astype(float)  # Clean and convert values

            # Scatter plot for the country
            plt.scatter(years, values, label=country, color=colors[i], marker=markers[i], s=100, edgecolor='k')

    # Customize the labels, title, and legend
    plt.xticks(rotation=90, fontsize=10)
    plt.yticks(fontsize=10)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Value', fontsize=12)
    plt.title('Scatter Plot of Countries Over the Years', fontsize=16, fontweight='bold')
    plt.legend(title="Countries", fontsize=10)
    
    # Add gridlines
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    
    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage for Canada, Costa Rica, Israel, Mexico, and India:
countries = ['Canada', 'Costa Rica', 'Israel', 'Mexico', 'India']
scatter_country_data(countries, df)
```

# Barplot for  Canada ,Costa Rica, Israel , Mexico ,India

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_country_data(country_name, df):
    # Filter the data for the specified country
    country_data = df[df['Geography 1/'] == country_name]

    # The years are column headers, so we need to transpose them for plotting
    years = country_data.columns[2:-1]  # Exclude the first two columns (continent and Geography) and the last NaN column
    values = country_data.iloc[0, 2:-1].replace(',', '', regex=True).astype(float)  # Clean values by removing commas and converting to float

    # Set the style and palette
    sns.set(style="whitegrid")
    colors = sns.color_palette("viridis", len(years))

    # Create the plot
    plt.figure(figsize=(12, 7))
    plt.bar(years, values, color=colors)

    # Customize the labels and title
    plt.xticks(rotation=90, fontsize=10, ha="center")
    plt.yticks(fontsize=10)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Value', fontsize=12)
    plt.title(f'{country_name} Data Over the Years', fontsize=16, fontweight='bold')

    # Add gridlines
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)

    # Tight layout to adjust for label rotation
    plt.tight_layout()

    # Show the plot
    plt.show()

# Example usage for Costa Rica, Israel, Mexico, and India:
plot_country_data('Canada', df)
plot_country_data('Costa Rica', df)
plot_country_data('Israel', df)
plot_country_data('Mexico', df)
plot_country_data('India', df)
plot_country_data('Brazil', df)
```

# piechart in 1989/90 of continents

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'C:/Users/User/cleaned_databook2.csv'
df = pd.read_csv(file_path)

# Clean the data by replacing commas and converting to numeric values
df_cleaned = df.copy()
df_cleaned.iloc[:, 2:-1] = df_cleaned.iloc[:, 2:-1].replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')

# Filter data for the year 1989/90
year = '1989/90'

# Group data by continent and sum values for 1989/90
continent_data = df_cleaned.groupby('continent')[year].sum()

# Set up color palette
colors = plt.cm.tab20.colors  # Using the tab20 color palette for better colors

# Create the pie chart
plt.figure(figsize=(10, 10))
wedges, texts, autotexts = plt.pie(
    continent_data,
    labels=continent_data.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=[0.1] * len(continent_data),  # Explode each slice for better visibility
    shadow=True,  # Add shadow to the pie chart
)

# Customize the text properties
for text in texts:
    text.set_fontsize(12)
    text.set_color('black')

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)

# Add a title
plt.title('Distribution of Total Values by Continent in 1989/90', fontsize=16, fontweight='bold')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')  

# Show the plot
plt.tight_layout()  # Adjust layout to make room for labels
plt.show()
```

# Forecasting 

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'C:/Users/User/cleaned_databook2.csv' 
df = pd.read_csv(file_path)

# Clean the data: Replace commas and convert to numeric
df_cleaned = df.copy()
df_cleaned.iloc[:, 2:-1] = df_cleaned.iloc[:, 2:-1].replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')

# Impute missing values with mean
df_cleaned.iloc[:, 2:-1] = df_cleaned.iloc[:, 2:-1].fillna(df_cleaned.iloc[:, 2:-1].mean())

# Focus on a specific country or continent for forecasting
# For example, if we want to forecast for 'Canada'
country_data = df_cleaned[df_cleaned['Geography 1/'] == 'Canada'].iloc[:, 2:-1].transpose()
country_data.columns = ['Value']  # Rename the column for clarity

# Ensure the values are numeric and not strings
country_data['Value'] = pd.to_numeric(country_data['Value'], errors='coerce')

# Drop any rows with NaN values (if any exist)
country_data = country_data.dropna()

# Convert index to datetime (if applicable)
country_data.index = pd.date_range(start='1989', periods=len(country_data), freq='A')  # Adjust frequency as needed

# Visualize the historical data
plt.figure(figsize=(10, 6))
plt.plot(country_data.index, country_data['Value'], marker='o')
plt.title('Historical Data for Canada')
plt.xlabel('Year')
plt.ylabel('Values')
plt.grid()
plt.show()

# Fit the ARIMA model
model = ARIMA(country_data['Value'], order=(1, 1, 1))  # Adjust (p, d, q) as necessary
model_fit = model.fit()

# Make predictions for the next 5 years
forecast = model_fit.forecast(steps=5)
forecast_index = pd.date_range(start=country_data.index[-1] + pd.DateOffset(years=1), periods=5, freq='A')

# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(country_data.index, country_data['Value'], label='Historical Data', marker='o')
plt.plot(forecast_index, forecast, label='Forecast', marker='x', color='red')
plt.title('Forecast for Canada using ARIMA')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()

# Print the forecasted values
print("Forecasted values for the next 5 years:")
for year, value in zip(forecast_index.year, forecast):
    print(f"{year}: {value:.2f}")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'C:/Users/User/cleaned_databook2.csv'
df = pd.read_csv(file_path)

# Clean the data: Replace commas and convert to numeric
df_cleaned = df.copy()
df_cleaned.iloc[:, 2:-1] = df_cleaned.iloc[:, 2:-1].replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')

# Impute missing values with mean
df_cleaned.iloc[:, 2:-1] = df_cleaned.iloc[:, 2:-1].fillna(df_cleaned.iloc[:, 2:-1].mean())

# Select data for a specific country (e.g., Canada) for forecasting
country_data = df_cleaned[df_cleaned['Geography 1/'] == 'Canada'].iloc[:, 2:-1].transpose()
country_data.columns = ['Value']
country_data['Value'] = pd.to_numeric(country_data['Value'], errors='coerce')
country_data = country_data.dropna()

# Convert index to datetime
country_data.index = pd.date_range(start='1989', periods=len(country_data), freq='A')

# Split the data into training and testing sets
train_size = int(len(country_data) * 0.8)  # Use 80% for training and 20% for testing
train_data, test_data = country_data[:train_size], country_data[train_size:]

# Fit the ARIMA model on the training data
model = ARIMA(train_data['Value'], order=(1, 1, 1))
model_fit = model.fit()

# Forecast the test period
forecast_test = model_fit.forecast(steps=len(test_data))
forecast_test = pd.Series(forecast_test, index=test_data.index)

# Calculate accuracy metrics
mse = mean_squared_error(test_data['Value'], forecast_test)
mae = mean_absolute_error(test_data['Value'], forecast_test)
mape = np.mean(np.abs((test_data['Value'] - forecast_test) / test_data['Value'])) * 100

print("Forecast Accuracy Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# Plot the historical data, test data, and forecast
plt.figure(figsize=(10, 6))
plt.plot(train_data.index, train_data['Value'], label='Training Data', marker='o')
plt.plot(test_data.index, test_data['Value'], label='Test Data', marker='o')
plt.plot(forecast_test.index, forecast_test, label='Forecast on Test Data', marker='x', color='red')
plt.title('Forecast Accuracy Testing for Canada using ARIMA')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()
```
